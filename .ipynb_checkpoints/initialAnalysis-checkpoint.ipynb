```python
def data_outliers(df):
    """
    Standarizing the data and any sample that has a value outside of the 3rd standard deviation
    is considered an outlier 
    
    Parameters
    ----------
    df : Pandas Dataframe
    
    Output
    ----------
    list : list of int. numbers indicating the indices of the outliers in the df
    """
    #standarizing
    sc = StandardScaler()
    sc.fit(df)
    df = pd.DataFrame(sc.transform(df))
    df = df[df<3] #Outside of the 3rd standard deviation 
    outliers_index = df[df.isnull().any(axis=1)].index.tolist()
    return outliers_index
outliers = data_outliers(df)
positiveOutliers = (df.iloc[outliers]['pH']<7.05).sum()
negativeOutliers = (df.iloc[outliers]['pH']>7.05).sum()
totalOutliers = len(outliers)
print('Total Outliers : %d, Positive Outliers : %d, Negative Outliers : %d' % (totalOutliers, positiveOutliers, negativeOutliers))
```

<h3>Logistic Regression</h3>

```python
df = df.drop(['diagnosis', 'segStage', 'segStageI_index', 'segStageII_index'], axis=1)
```

```python
#______________ENCODING THE PH COLUMN____________________
# 0: healthy, 1: positive(acidosis)
df['pH'] = (df.pH <= 7.05).astype(int).ravel()
df.shape #--> (3404, 48)
```

```python
y = df['pH'] # target
X = df.drop(['pH'], axis=1) # Feature matrix
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
#standarizing the training data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
columns = list(X_train)
X_train = pd.DataFrame(sc.transform(X_train))
X_train.columns = columns
```

```python
from sklearn.linear_model import LogisticRegression
lr_rigid = LogisticRegression(solver='lbfgs', penalty='l2', random_state=1)
lr_lasso = LogisticRegression(penalty='l1', random_state=1)
lr_rigid.fit(X_train, y_train)
lr_lasso.fit(X_train, y_train)
```

```python
X_test = sc.transform(X_test)
y_pred_class_rigid = lr_rigid.predict(X_test)
y_pred_class_lasso = lr_lasso.predict(X_test)
```

### Performance 

#### F1 Score and precision :

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score
roc_auc_lasso = cross_val_score(lr_lasso, X, y, cv=10, scoring='roc_auc')
roc_auc_rigid = cross_val_score(lr_rigid, X, y, cv=10, scoring='roc_auc')
f1_lasso = cross_val_score(lr_lasso, X, y, cv=10, scoring='f1')
f1_rigid = cross_val_score(lr_rigid, X, y, cv=10, scoring='f1')
precision_lasso = cross_val_score(lr_lasso, X, y, cv=10, scoring='precision')
precision_rigid = cross_val_score(lr_rigid, X, y, cv=10, scoring='precision')
print('######################################\n')
print('                 Lasso          rigid \n')
print('roc_auc   :      %f             %f  \n' % (roc_auc_lasso.mean(), roc_auc_rigid.mean()))
print('f1        :      %f             %f  \n' % (f1_lasso.mean(), f1_rigid.mean()))
print('precision :      %f             %f  \n' % (precision_lasso.mean(), precision_rigid.mean()))
print('######################################\n')
```

```python
# For L2 penalty
confusion_l2 = metrics.confusion_matrix(y_test, y_pred_class_rigid)
# For L1 penalty
confusion_l1 = metrics.confusion_matrix(y_test, y_pred_class_lasso)
```

```python
class_names=[0,1]
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
plt.subplot(2,1,1)
sns.heatmap(pd.DataFrame(confusion_l2), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix for l2 reg.', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.subplot(2,1,2)
sns.heatmap(pd.DataFrame(confusion_l1), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix for l1 reg.', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

<h4>ROC AND AUC </h4>

```python
y_proba_l2 = lr_rigid.predict_proba(X_test)[:, 1]
y_proba_l1= lr_lasso.predict_proba(X_test)[:, 1]
fpr_l2, tpr_l2, threshold_l2 = metrics.roc_curve(y_test, y_proba_l2)
fpr_l1, tpr_l1, threshold_l1 = metrics.roc_curve(y_test, y_proba_l1)
```

```python
plt.plot(fpr_l2, tpr_l2, label='logre.l2 = %0.2f' % metrics.auc(fpr_l2, tpr_l2))
plt.plot(fpr_l1, tpr_l1, label='logre.l1 = %0.2f' % metrics.auc(fpr_l1, tpr_l1))
plt.plot([0, 1], [0, 1], color='navy', lw=0.5, linestyle='--')
plt.legend(loc = 'lower right')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC Curve for different classifiers')
plt.xlabel('FP rate')
plt.ylabel('TP rate')
plt.grid(True)
```

```python
def plot_regularization(X_train, columns):
    """
    ploting the strength of the regularization parameter C vs the weights
    
    Parameters
    ----------
    X_train : Pandas Dataframe, training dataset
    
    Output
    ----------
    plot : C vs weights
"""
    
    weights, params = [], []
    columns = X_train.columns
    for c in np.arange(-5, 5):
        lr = LogisticRegression(solver='lbfgs', C=10.**c, random_state=1)
        lr.fit(X_train, y_train)
        weights.append(lr.coef_[0])
        params.append(10.**c)
    weights = np.array(weights)
    plt.figure(figsize=(18,15))
    for i in range(len(columns)):
        plt.plot(params, weights[:, i], label=columns[i])
    plt.ylabel('weight coefficient')
    plt.xlabel('C')
    plt.legend(loc='upper left')
    plt.xscale('log')
    plt.grid(True)
    plt.show()
```

#### Roc Curve for different parameter C (threshold) in the logistic regression

```python

a_l1, a_l2, par= [], [], []#AUC score for different parameter C a_l1, 
for c in np.arange(-6, 6):
    lr_l1 = LogisticRegression(penalty='l1', C=10.**c, random_state=1)
    lr_l2 = LogisticRegression(solver='lbfgs', penalty='l2', C=10.**c, random_state=1)
    lr_l1.fit(X_train, y_train)
    lr_l2.fit(X_train, y_train)
    y_proba_l1 = lr_l1.predict_proba(X_test)[:, 1]
    y_proba_l2 = lr_l2.predict_proba(X_test)[:, 1]
    fpr_l1, tpr_l1, _ = metrics.roc_curve(y_test, y_proba_l1)
    fpr_l2, tpr_l2, _ = metrics.roc_curve(y_test, y_proba_l2)
    auc_l1 = metrics.auc(fpr_l1, tpr_l1)
    auc_l2 = metrics.auc(fpr_l2, tpr_l2)
    a_l1.append(auc_l1)
    a_l2.append(auc_l2)
    par.append(c)
plt.ylabel('AUC')
plt.xlabel('Log C')
plt.grid(True)
plt.plot(par, a_l1)
plt.plot(par, a_l2)
plt.legend()
```
